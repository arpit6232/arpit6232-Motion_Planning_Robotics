```python
# Global Flags
DSTAR  = 2
ATTRACT_GAIN = 1/50
REPULSIVE_GAIN = 200
REPULSIVE_QSTAR = 2
POTENTIAL_WIDTH = 5.0  # potential area width [m]
MAX_ITERATIONS = 700

PIXEL_MIN = 0
PIXEL_MAX = 160

RANGE_MIN = -20
RANGE_MAX = 20

plt.xlim([0, 160])
plt.ylim([0, 160])
```

```python
def range_to_pixel(val):
    return (((val - RANGE_MIN) * (PIXEL_MAX - PIXEL_MIN)) / (RANGE_MAX-RANGE_MIN)) + PIXEL_MIN

def pixel_to_range(val):
    return (((val - PIXEL_MIN) * (RANGE_MAX-RANGE_MIN)) / (PIXEL_MAX - PIXEL_MIN)) + RANGE_MIN
```

```python
def attractive_potential_func(x, y, q_start, q_goal):
    attractive = 0.5 * ATTRACT_GAIN  * ( (x - q_goal[0])**2 + (y - q_goal[1])**2 )
    return attractive
```

```python
def repulsive_potential(obstacle):
    distance_where_pixel_Empty = bwdist(obstacle==0)

    d2 = (distance_where_pixel_Empty/100.) + 1
    repulsive = 0.5* REPULSIVE_GAIN *((1./d2 - 1/REPULSIVE_QSTAR)**2)
    repulsive [d2 > REPULSIVE_QSTAR] = 0

    # return repulsive / np.linalg.norm(repulsive)
    return repulsive
```

```python
def calc_potential_field(q_start, q_goal, obstacle, grid_size):
    q_start = [range_to_pixel(0.0), range_to_pixel(0.0)]
    q_goal = [range_to_pixel(10.0), range_to_pixel(0.0)]

    x_hat = np.arange(0, max(20, 160), 1) ############CHANGE THIS
    y_hat = np.arange(0, max(20, 160), 1) ############CHANGE THIS

    
    [x, y] = np.meshgrid (x_hat, y_hat)

    obstacle = np.zeros((x.shape[0],x.shape[1]))

    # # W = WORKSPACE_CONFIG['WO3']
    # # obst = pc.Region(list_poly=W)

    obstacle[ 94:98, 82:86 ] = True
    obstacle[ 106:110, 74:78 ] = True

    # X, Y = np.meshgrid(x, y)
    repulsive = repulsive_potential(obstacle)
    attractive = attractive_potential_func(x, y, q_start,  q_goal)
    U = repulsive + attractive

    return U
```

```python
def GradientBasedPlanner (f, start_coords, end_coords, max_its):
    # GradientBasedPlanner : This function plans a path through a 2D
    # environment from a start to a destination based on the gradient of the
    # function f which is passed in as a 2D array. The two arguments
    # start_coords and end_coords denote the coordinates of the start and end
    # positions respectively in the array while max_its indicates an upper
    # bound on the number of iterations that the system can use before giving
    # up.
    # The output, route, is an array with 2 columns and n rows where the rows
    # correspond to the coordinates of the robot as it moves along the route.
    # The first column corresponds to the x coordinate and the second to the y coordinate

    [gy, gx] = np.gradient(-U);

    route = np.vstack( [np.array(start_coords), np.array(start_coords)] )
    for i in range(max_its):
        current_point = route[-1,:];
#         print(sum( abs(current_point-end_coords) ))
        if sum( abs(current_point-end_coords) ) < 2:
            print('Reached the goal !');
            break
        ix = int(round( current_point[1] ));
        iy = int(round( current_point[0] ));
        vx = gx[ix, iy]
        vy = gy[ix, iy]
        dt = 1 / np.linalg.norm([vx, vy]);
        next_point = current_point + dt*np.array( [vx, vy] );
        route = np.vstack( [route, next_point] );
    route = route[1:,:]
        
    return route
```
